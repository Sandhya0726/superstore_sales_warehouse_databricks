### Sales Pipeline

#### Data Ingestion

```python
df = spark.read.format("csv") \
    .option("header", True) \
    .option("inferSchema", True) \
    .option("multiLine", True) \
    .option("quote", "\"") \
    .option("escape", "\"") \
    .load("/Volumes/workspace/default/practice/SuperStore_Sales/SuperStoreOrders - SuperStoreOrders.csv")
display(df)
```

#### store the raw data into silver layer in delta format

```python
from pyspark.sql.functions import current_date
```

#### Adding metadata, current_date to make it easy for audit and debugging

```python
df=df.withColumn('ingestion_date',current_date())
```

```python
df.write.format('delta').mode('overwrite').option('overwriteSchema',True).save('/Volumes/workspace/default/practice/SuperStore_Sales/Bronze')
```
